```python

# -*- coding: utf-8 -*-
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# Matplotlib中文显示设置（确保中文标签正常显示）
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置字体为黑体SimHei
plt.rcParams['axes.unicode_minus'] = False    # 正常显示负号

# 1. 数据生成与初步预处理
np.random.seed(42)  # 固定随机种子，确保结果可复现
num_users = 1000
# 随机生成用户属性：性别、年龄、活跃度、地区
genders = np.random.choice(['Male', 'Female'], size=num_users)
ages = np.random.randint(18, 61, size=num_users)  # 18到60岁的随机年龄
activity_levels = np.random.choice(['Sedentary', 'Moderate', 'Active'],
                                   p=[0.2, 0.5, 0.3], size=num_users)
locations = np.random.choice(['Urban', 'Suburban', 'Rural'],
                             p=[0.5, 0.3, 0.2], size=num_users)
# 根据活跃度调整年度使用次数：久坐用户更少，活跃用户更多
base_sessions = np.random.normal(50, 15, num_users)    # 初始年度使用次数（均值50，标准差15）
base_sessions[activity_levels == 'Sedentary'] *= 0.5   # 久坐(Sedentary)用户次数减半
base_sessions[activity_levels == 'Active'] *= 1.5      # 高度活跃(Active)用户次数增加50%
app_sessions = np.clip(base_sessions, 5, None).astype(int)  # 下限为5次并取整

# 模拟运动距离(公里)和卡路里消耗：假设每次使用平均行驶0.5公里，每公里消耗约10卡路里
distance = (app_sessions * np.random.normal(0.5, 0.1, num_users)).round(1)  # 总行驶距离
distance = np.clip(distance, 0, None)  # 距离不为负值
calories = (distance * 10 + np.random.normal(0, 50, num_users)).astype(int)  # 总卡路里消耗
calories = np.clip(calories, 0, None)  # 卡路里不为负值

# 整合生成DataFrame
data = pd.DataFrame({
    'Gender': genders,
    'Age': ages,
    'ActivityLevel': activity_levels,
    'Location': locations,
    'Sessions': app_sessions,
    'DistanceKM': distance,
    'Calories': calories
})

# 衍生特征：平均每次使用的距离和卡路里（总距离/总次数， 总卡路里/总次数）
data['Distance_per_Session'] = (data['DistanceKM'] / data['Sessions']).round(2)
data['Calories_per_Session'] = (data['Calories'] / data['Sessions']).round(2)

# 2. 探索性数据分析（EDA）
# 描述性统计：查看主要数值特征的集中趋势和分散程度
desc_stats = data[['Sessions', 'DistanceKM', 'Calories',
                   'Distance_per_Session', 'Calories_per_Session', 'Age']].describe()
print("数据概览：\n", desc_stats, "\n")

# 绘制年度使用次数的分布直方图
plt.figure(figsize=(6,4))
plt.hist(data['Sessions'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel("年度使用次数"); plt.ylabel("用户数量")
plt.title("App 使用次数分布图")
plt.show()

# 绘制数值变量之间的相关性热力图
plt.figure(figsize=(6,5))
corr_matrix = data[['Sessions', 'DistanceKM', 'Calories', 'Age',
                    'Distance_per_Session', 'Calories_per_Session']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu')
plt.title("用户指标相关性热力图")
plt.show()

# 绘制距离与卡路里散点图，并添加线性趋势线
plt.figure(figsize=(6,4))
sns.regplot(x='DistanceKM', y='Calories', data=data, line_kws={'color': 'red'})
plt.xlabel("运动距离 (公里)"); plt.ylabel("卡路里消耗")
plt.title("距离与卡路里消耗关系 (含趋势线)")
plt.show()

# 绘制不同活跃度用户的年度使用次数箱线图
plt.figure(figsize=(6,4))
sns.boxplot(x='ActivityLevel', y='Sessions', data=data, palette='Pastel1')
plt.xlabel("活跃度等级"); plt.ylabel("年度使用次数")
plt.title("不同活跃度用户的使用次数分布")
plt.show()

# 3. 假设检验 (Hypothesis Testing)
from scipy.stats import ttest_ind, f_oneway, chi2_contingency

# t检验：比较男性和女性用户年度使用次数的差异是否显著
male_sessions = data[data['Gender'] == 'Male']['Sessions']
female_sessions = data[data['Gender'] == 'Female']['Sessions']
t_stat, p_gender = ttest_ind(male_sessions, female_sessions, equal_var=False)  # Welch’s t检验
mean_male = male_sessions.mean(); mean_female = female_sessions.mean()
print(f"男性平均使用次数: {mean_male:.1f}, 女性平均使用次数: {mean_female:.1f}")
print(f"性别差异检验 t检验 p值: {p_gender:.3f}")
if p_gender < 0.05:
    print("结论: 男性与女性的使用次数存在显著差异\n")
else:
    print("结论: 男性与女性的使用次数无显著差异\n")

# 单因素方差分析 ANOVA：不同活跃度用户的使用次数是否有显著差异
sedentary = data[data['ActivityLevel'] == 'Sedentary']['Sessions']
moderate  = data[data['ActivityLevel'] == 'Moderate']['Sessions']
active    = data[data['ActivityLevel'] == 'Active']['Sessions']
f_stat, p_activity = f_oneway(sedentary, moderate, active)
means = data.groupby('ActivityLevel')['Sessions'].mean()
print("不同活跃度组平均使用次数:",
      f"Sedentary: {means['Sedentary']:.1f}, Moderate: {means['Moderate']:.1f}, Active: {means['Active']:.1f}")
print(f"活跃度差异检验 ANOVA p值: {p_activity:.3e}")
if p_activity < 0.05:
    print("结论: 不同活跃度用户的使用次数存在显著差异\n")
else:
    print("结论: 不同活跃度用户的使用次数无显著差异\n")

# 卡方检验：性别与用户高参与度(Engaged)是否有关联
data['Engaged'] = (data['Sessions'] >= 70).astype(int)  # 定义高参与度用户（年度使用次数>=70次）
contingency = pd.crosstab(data['Gender'], data['Engaged'])
chi2, p_chi, dof, expected = chi2_contingency(contingency)
eng_rate_male = data[data['Gender'] == 'Male']['Engaged'].mean()
eng_rate_female = data[data['Gender'] == 'Female']['Engaged'].mean()
print(f"男性用户参与率: {eng_rate_male*100:.1f}%, 女性用户参与率: {eng_rate_female*100:.1f}%")
print(f"性别与参与度独立性检验 卡方检验 p值: {p_chi:.3f}")
if p_chi < 0.05:
    print("结论: 性别与用户参与度存在显著关联\n")
else:
    print("结论: 性别与用户参与度无显著关联\n")

# 4. 特征工程 (Feature Engineering)
# 标准化主要数值特征，以消除量纲影响
numeric_features = ['Sessions', 'DistanceKM', 'Calories', 'Age',
                    'Distance_per_Session', 'Calories_per_Session']
scaler = StandardScaler()
data_scaled = data.copy()
data_scaled[numeric_features] = scaler.fit_transform(data_scaled[numeric_features])

# 利用DBSCAN聚类进行异常值检测（识别行为异常的用户）
from sklearn.cluster import DBSCAN
X_usage = data_scaled[['Sessions', 'DistanceKM', 'Calories']]  # 选择主要使用行为特征
dbscan = DBSCAN(eps=0.5, min_samples=5).fit(X_usage)
outlier_indices = np.where(dbscan.labels_ == -1)[0]
print(f"DBSCAN检测到 {len(outlier_indices)} 个异常用户 (共 {num_users} 名用户)")
if len(outlier_indices) > 0:
    print("异常用户索引:", outlier_indices, "\n")
else:
    print("无明显异常值\n")

# 5. 机器学习模型对比 (Logistic Regression vs Random Forest)
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# 构造特征矩阵X和目标变量y（预测用户是否为高参与度用户）
# 为避免目标泄漏，不将Sessions直接作为特征，因为Engaged由Sessions定义
features = ['DistanceKM', 'Calories', 'Age', 'Gender', 'ActivityLevel',
            'Location', 'Distance_per_Session', 'Calories_per_Session']
X = data[features].copy()
y = data['Engaged']

# 将分类变量进行独热编码，去除一个基准列避免多重共线
X = pd.get_dummies(X, columns=['Gender', 'ActivityLevel', 'Location'], drop_first=True)

# 划分训练集和测试集（70%训练集，30%测试集）
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 对数值特征进行标准化（基于训练集均值和标准差）
num_cols = ['DistanceKM', 'Calories', 'Age', 'Distance_per_Session', 'Calories_per_Session']
scaler2 = StandardScaler()
X_train[num_cols] = scaler2.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler2.transform(X_test[num_cols])

# 初始化模型：逻辑回归与随机森林
lr_model = LogisticRegression(max_iter=500, random_state=42)
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# 使用5折交叉验证评估模型在训练集上的表现（准确率）
lr_scores = cross_val_score(lr_model, X_train, y_train, cv=5, scoring='accuracy')
rf_scores = cross_val_score(rf_model, X_train, y_train, cv=5, scoring='accuracy')
print(f"逻辑回归模型 5折交叉验证平均准确率: {lr_scores.mean()*100:.2f}%")
print(f"随机森林模型 5折交叉验证平均准确率: {rf_scores.mean()*100:.2f}%")

# 在训练集上训练模型，并在测试集上评估模型表现
lr_model.fit(X_train, y_train)
rf_model.fit(X_train, y_train)
lr_acc = lr_model.score(X_test, y_test)
rf_acc = rf_model.score(X_test, y_test)
print(f"逻辑回归模型 测试集准确率: {lr_acc*100:.2f}%")
print(f"随机森林模型 测试集准确率: {rf_acc*100:.2f}%")
```
